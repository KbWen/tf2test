```python
%matplotlib widget
import matplotlib.pyplot as plt
import tensorflow as tf
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

# check gpu
tf.config.list_physical_devices()
tf.test.is_built_with_cuda()
```

```python
df_train = pd.read_csv("./data_set/house_prices/train.csv")
df_train.head()
```

```python
df_train.info()
```

```python
df_train['SalePrice'].describe()
```

```python
df_test = pd.read_csv("./data_set/house_prices/test.csv")

df_train.drop(['Id'], axis=1, inplace=True)
df_test.drop(['Id'], axis=1, inplace=True)

df_train.reset_index(drop=True, inplace=True)
df_train["SalePrice"] = np.log1p(df_train["SalePrice"])
y = df_train['SalePrice'].reset_index(drop=True)

train_features = df_train.drop(['SalePrice'], axis=1)
test_features = df_test
features = pd.concat([train_features, test_features]).reset_index(drop=True)
```

```python
# features.shape
features.head()
# df_train.head()
```

```python
df_train['is_train'] = 1
df_test['is_train'] = 0

df_data = pd.concat([df_train, df_test])

df_data.fillna(0, inplace=True)

df_data.head()

categorical_features = df_data.select_dtypes(include=['object'])
categorical_features_one_hot = pd.get_dummies(categorical_features)
categorical_features_one_hot.head()

data = pd.concat([df_data, categorical_features_one_hot], axis=1)
data.drop(categorical_features, axis=1, inplace=True)
```

```python
prepro_data = MinMaxScaler()
col = list(data.columns)
data = pd.DataFrame(prepro_data.fit_transform(data), columns = col)
```

```python
data.head()
```

```python
train = data[data['is_train'] == 1]
test = data[data['is_train'] == 0]

train = train.drop(['is_train'], axis=1)
train = train.drop(['SalePrice'], axis=1)
y = df_train["SalePrice"]

test = test.drop(['is_train'], axis=1)
test = test.drop(['SalePrice'], axis=1)
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(train ,y , test_size=0.2, random_state=42)
```

```python
model = tf.keras.Sequential()

model.add(tf.keras.layers.Dense(256))
model.add(tf.keras.layers.Activation(tf.keras.layers.LeakyReLU(alpha=0.1)))
model.add(tf.keras.layers.Dropout(0.8))

model.add(tf.keras.layers.Dense(128))
model.add(tf.keras.layers.Activation('relu'))
model.add(tf.keras.layers.Dropout(0.8))

model.add(tf.keras.layers.Dense(64))
model.add(tf.keras.layers.Activation('relu'))
model.add(tf.keras.layers.Dropout(0.8))

model.add(tf.keras.layers.Dense(16))
model.add(tf.keras.layers.Activation('relu'))
model.add(tf.keras.layers.Dropout(0.8))

model.add(tf.keras.layers.Dense(1))

model.compile(loss='mean_squared_error', metrics=['mse'], optimizer='adam')
```

```python
history = model.fit(X_train, y_train, epochs=1000, validation_split=0.15)
```

```python
y_pred = model.predict(test)
```

```python
y_pred = np.expm1(y_pred)
```

```python
y_pred = y_pred.reshape(1459,)
```

```python
df_test_bug = pd.read_csv("./data_set/house_prices/test.csv")

submission_df = pd.DataFrame({'Id':df_test_bug['Id'], 'SalePrice':y_pred})
submission_df.to_csv('Submission.csv', index=False)
```
